### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ventas_tienda = tienda['Precio'].sum()
print(ventas_tienda)
ventas_tienda2 = tienda2['Precio'].sum()
print(ventas_tienda2)
ventas_tienda3 = tienda3['Precio'].sum()
print(ventas_tienda3)
ventas_tienda4 = tienda4['Precio'].sum()
print(ventas_tienda4)
```

```python
import matplotlib.pyplot as plt

ventas = [ventas_tienda, ventas_tienda2, ventas_tienda3, ventas_tienda4]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

plt.figure(figsize=(6,6))
plt.pie(ventas, labels=tiendas, autopct='%1.1f%%')
plt.title("Participación de ventas por tienda")
plt.show()
```

# 2. Ventas por categoría

```python
ventas_por_categoria_tienda = tienda.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)
print(ventas_por_categoria_tienda)
ventas_por_categoria_tienda2 = tienda2.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)
print(ventas_por_categoria_tienda2)
ventas_por_categoria_tienda3 = tienda3.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)
print(ventas_por_categoria_tienda3)
ventas_por_categoria_tienda4 = tienda4.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)
print(ventas_por_categoria_tienda4)
```

# 3. Calificación promedio de la tienda


```python
calificacion_tienda = tienda[['Calificación']].mean()
print(calificacion_tienda)
calificacion_tienda2 = tienda2[['Calificación']].mean()
print(calificacion_tienda2)
calificacion_tienda3 = tienda3[['Calificación']].mean()
print(calificacion_tienda3)
calificacion_tienda4 = tienda4[['Calificación']].mean()
print(calificacion_tienda4)
```

# 4. Productos más y menos vendidos

```python
print('Productos mas vendidos')

top_5_tienda = tienda.Producto.value_counts().head(50)
print (top_5_tienda)
top_5_tienda2 = tienda2.Producto.value_counts().head()
print (top_5_tienda2)
top_5_tienda3 = tienda3.Producto.value_counts().head()
print (top_5_tienda3)
top_5_tienda4 = tienda4.Producto.value_counts().head()
print (top_5_tienda4)

print('Productos menos vendidos')
top_5_tienda_peores = tienda.Producto.value_counts().tail(5)
print (top_5_tienda_peores)
top_5_tienda2_peores = tienda2.Producto.value_counts().tail(5)
print (top_5_tienda2_peores)
top_5_tienda3_peores = tienda3.Producto.value_counts().tail(5)
print (top_5_tienda3_peores)
top_5_tienda4_peores = tienda4.Producto.value_counts().tail(5)
print (top_5_tienda4_peores)
```

```python
top_suma_tienda = tienda.Producto.value_counts().sum()
print (top_suma_tienda)

top_suma_tienda2 = tienda2.Producto.value_counts().sum()
print (top_suma_tienda2)

top_suma_tienda3 = tienda3.Producto.value_counts().sum()
print (top_suma_tienda3)

top_suma_tienda4 = tienda4.Producto.value_counts().sum()
print (top_suma_tienda4)
```

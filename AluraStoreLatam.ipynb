### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ventas_tienda = tienda['Precio'].sum()
print(ventas_tienda)
ventas_tienda2 = tienda2['Precio'].sum()
print(ventas_tienda2)
ventas_tienda3 = tienda3['Precio'].sum()
print(ventas_tienda3)
ventas_tienda4 = tienda4['Precio'].sum()
print(ventas_tienda4)
```

```python
import matplotlib.pyplot as plt

ventas = [ventas_tienda, ventas_tienda2, ventas_tienda3, ventas_tienda4]
tiendas = ['Tienda1', 'Tienda2', 'Tienda3', 'Tienda4']

plt.figure(figsize=(8,5))
plt.bar(tiendas, ventas)
plt.xlabel('Tiendas')
plt.ylabel('Ventas')
plt.title('Ventas por tienda')
plt.show()
```

# 2. Ventas por categoría

```python
ventas_por_categoria_tienda = tienda.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)
print(ventas_por_categoria_tienda)
ventas_por_categoria_tienda2 = tienda2.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)
print(ventas_por_categoria_tienda2)
ventas_por_categoria_tienda3 = tienda3.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)
print(ventas_por_categoria_tienda3)
ventas_por_categoria_tienda4 = tienda4.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)
print(ventas_por_categoria_tienda4)
```

# 3. Calificación promedio de la tienda


```python
calificacion_tienda = tienda[['Calificación']].mean()
print(calificacion_tienda)
calificacion_tienda2 = tienda2[['Calificación']].mean()
print(calificacion_tienda2)
calificacion_tienda3 = tienda3[['Calificación']].mean()
print(calificacion_tienda3)
calificacion_tienda4 = tienda4[['Calificación']].mean()
print(calificacion_tienda4)
```

# 4. Productos más y menos vendidos

```python
top_5_tienda = tienda.Producto.value_counts().head(5)
print (top_5_tienda)
top_5_tienda2 = tienda2.Producto.value_counts().head(5)
print (top_5_tienda2)
top_5_tienda3 = tienda3.Producto.value_counts().head(5)
print (top_5_tienda3)
top_5_tienda4 = tienda4.Producto.value_counts().head(5)
print (top_5_tienda4)
```

# 5. Envío promedio por tienda

```python
costo_de_envio_tienda = tienda[['Costo de envío']].mean()
print(costo_de_envio_tienda)
costo_de_envio_tienda2 = tienda2[['Costo de envío']].mean()
print(costo_de_envio_tienda2)
costo_de_envio_tienda3 = tienda3[['Costo de envío']].mean()
print(costo_de_envio_tienda3)
costo_de_envio_tienda4 = tienda4[['Costo de envío']].mean()
print(costo_de_envio_tienda4)
```
